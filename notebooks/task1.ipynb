```python
# Calculate IQR
Q1 = df['TotalClaims'].quantile(0.25)
Q3 = df['TotalClaims'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df[(df['TotalClaims'] < lower_bound) | (df['TotalClaims'] > upper_bound)]
print(outliers)
```

```python
overall_loss_ratio = df['TotalClaims'].sum() / df['TotalPremium'].sum()
print(overall_loss_ratio)
```
